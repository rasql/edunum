# Sous-graphes

Une figure peut être découpé en multiples sous-graphes (subplots).

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Subplots
La fonction `subplot(n, m)` crée $n \times m$ sous-graphes.

```python
plt.subplots();
```

```python
plt.subplots(2, 1);
```

```python
plt.subplots(1, 2);
```

## Sélectionner un sous-graphe
La fonction `subplot(n, m, i)` sélectionne le i-ième sous-graphique d'une matrice de $n \times m$ sous-graphes. 

Si le nombre des graphes est inférieur à 10 on peut les désigner directement avec `subplot(nmi)`

```python
x = [0, 1, 2, 3, 4]
y = [1, 3, 2, 5, 2]
```

```python
plt.subplot(121)
plt.plot(y, 'ro-')
plt.subplot(122)
plt.bar(x, y);
```

## Graphes superposés

```python
t = np.arange(0.0, 5.0, 0.1)

plt.subplot(211)
plt.plot(t, np.exp(-t) * np.cos(2*np.pi*t), 'bo-')

plt.subplot(212)
plt.plot(t, np.cos(2*np.pi*t), 'r--');
```

## Graphes en 2x2

Nous pouvons aussi sous-diviser en $2 \times 2$ sous-graphes.

```python
plt.subplot(221)
plt.bar(x, y)
plt.subplot(222)
plt.plot(y, 'ro-');
plt.subplot(223)
plt.plot(x, y)
plt.subplot(224)
plt.plot(x, y, 'og');
```

## Arrangement irrégulier

Il est possible de varier le nombre de sous-graphes dans chaque ligne.

```python
plt.subplot(2,1,1)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,4)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,5)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,6)
plt.xticks([]), plt.yticks([]);
```
