# Nuage de points

Le graphique en nuage de points (scatter plot en anglais) affiche des points  de taille diff√©rents.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 50
x, y = np.random.randn(2, n)
r = np.random.randint(10, 200, n)
plt.scatter(x, y, r, c=r);
```

```python
n = 50
x, y = np.random.randn(2, n)
s = np.random.randint(10, 300, n)
plt.scatter(x, y, s, s, cmap='hot');
plt.colorbar();
```

## Nuage de points


```python
n = 256
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=200, c=T, alpha=.5)

#plt.xlim(-1.5,1.5), plt.xticks([])
#plt.ylim(-1.5,1.5), plt.yticks([])
plt.savefig('scatter_ex.png',dpi=48)
plt.show()
```

Les coordonn√©es x et y sont suivent une distribution normale avec moyenne=0 et sigma=1.

```python
plt.plot(X, label='x')
plt.plot(Y, 'r', label='y')
plt.legend();
```

Les positions se groupent autour de l'orignine (0, 0).

```python
plt.plot(X, Y, 'o-');
```

L'arctangente `arctan2` donne des angles dans l'intervalle $[-\pi, +\pi]$

```python
plt.plot(T, 'o-');
```
