# Graphes

Le module `matplotlib` permet d'afficher des graphes. Le module `numpy` permet de faire du calcul numérique avec des matrices. 

Pour utiliser ces modules nous devons les importer. Il est tradition de les importer sous les noms raccourcis suivants.

- `plt` pour **pyplot**
- `np` pour **numpy**

```python
import matplotlib.pyplot as plt
import numpy as np
```

Créons une liste de données que nous pouvons afficher.

```python
y = [2, 4, 3, 6, 3]
```

Un graphique simple (plot en anglais) est obtenu avec ceci.

```python
plt.plot(y)
```

L'objet retourné est de type `Line2D`.
Dorénavant nous terminons la dernière ligne avec un point-virgule `;` pour supprimer cette information.

**Attention:**  
Pour afficher un graphique dans votre programme Python, vous devez ajouter la commande `plt.show()` pour l'afficher dans une fenêtre à part.  Dans ce document-ci cette commande est inclut de façon implicite.

## Titres et légendes
Nous pouvons ajouter un **titre** ainsi que des **légendes** pour les deux axes x et y.

```python
plt.title('Evolution')
plt.xlabel('temps')
plt.ylabel('montant')
plt.plot(y);
```

## Style des lignes
Nous pouvons également ajouter un string de formatage. Les couleurs sont indiqué par
- `r` rouge
- `g` vert (green)
- `b` bleu
- `y` jaune (yellow)
- `m` magenta
- `c` cyan
- `k` noir (black)
- `w` blanc (white)

```python
plt.plot(y, 'mo-');
```

La forme des points est indiqué par
- `o` cercle
- `s` carré (square)
- `^` triangle (haut)
- `v` triangle (bas)
- `*` asterisque
- `x` x
- `+` plus

```python
plt.plot(y, 'gv:');
```

Le format de ligne est indiqué par
- `-` solide
- `:` pointillé
- `--` traitille
- `-.` mixte


```python
plt.plot(y, 'cs-.');
```

Nous pouvons mélanger deux styles. Pour ceci nous devons afficher deux lignes.

```python
plt.plot(y, 'b:', y, 'or')
```

## Coordonnées dans x

Si nous affichons un graphe à partir d'un seul vecteur, les valeurs x sont sont des entiers commençant à 0. Nous pouvons définir notre propre plage x, par exemple avec un vecteur **x** allant de -2 à +2.

```python
x = [-2, -1, 0, 1, 2]
plt.plot(x, y,'k^:');
```

## Propriétés de ligne

Avec des arguments à options nous pouvons changer les propriétés d'une ligne.

```python
plt.plot(x, y, color='r', 
         linewidth=2, linestyle='--', 
         marker='o', markersize=12, 
         markerfacecolor='yellow', 
         markeredgewidth=2, markeredgecolor='blue');
```

## Multiples lignes

Nous pouvons placer multiples lignes sur un plot.

```python
t = np.arange(0., 2., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs:', t, t**3, 'g^:')
```

## Courbes trigonométriques

Le module **numpy** nous permet l'accès faciles aux fonctions trigonométriques.

```python
t = np.linspace(0, 4*np.pi, 100)
plt.plot(t, np.sin(t), label='sin(t)')
plt.plot(t, np.cos(t), label='cos(t)')
plt.legend()
plt.grid()
```

Ici nous affichons le carré du sinus.

$$ f(t) = (\sin t)^2 $$

```python
t = np.linspace(0, 4*np.pi, 100)
plt.plot(t, np.sin(t), label='sin(t)')
plt.plot(t, np.sin(t)**2, label='sin(t)**2')
plt.legend()
plt.grid()
```

## Figures de Lissajous

Les figures des Lissajous sont des combinaisons de courbes sinus et cosinus.

```python
t = np.linspace(0, 2*np.pi, 100)
plt.grid()
plt.plot(t, np.sin(3*t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(3*t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(2*t), np.cos(3*t), 'r');
```

## Echelle logarithmiques

Nous pouvons afficher une axe

```python
y = np.arange(0, 100, 2)
plt.grid()
plt.plot(y, 'o-')
plt.yscale('log')
```

```python
plt.grid()
plt.plot(y, 'o-')
plt.xscale('log')
```

```python
plt.grid()
plt.plot(y, 'o')
plt.xscale('log')
plt.yscale('log')
```
