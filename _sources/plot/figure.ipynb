# Figure et Axes

Un graphique consiste des objets

- `Figure` l'espace total du graphique
- `Axes` la partie qui présente les données

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Figure

La fonction `subplots()` retourne un tuple consistant de

- `Figure` le conteneur global
- `Axes` pour présenter les données

```python
fig, ax = plt.subplots()
```

La figure a environs 150 methodes.

```python
f = [x for x in dir(fig) if not x.startswith('_')]
print(len(f))
f[:20]
```

## Taille

La taille de la figure par défaut est 6 x 4 pouces.

```python
fig.get_figwidth(), fig.get_figheight()
```

Nous pouvons modifier la taille avec le mot-clé `figsize`.

```python
plt.subplots(figsize=(8, 2));
```

## Couleur
La couleur du graphique est choisi avec `facecolor`.

```python
plt.subplots(facecolor='lightgray', figsize=(8, 2));
```

La couleur du contour et son épaisseur peuvent être choisis.

```python
plt.subplots(figsize=(8, 2), facecolor='pink', 
             edgecolor='red', linewidth=3, frameon=True);
```

```python
y = [2, 4, 3, 5, 1]
plt.subplots(figsize=(8, 3), facecolor='pink', 
             edgecolor='red', linewidth=3, frameon=True)
plt.plot(y, 'or-');
```

## Ligne

Des lignes peuvent être ajoutés.

```python
plt.plot(y);
plt.axhline(2, c='r')
plt.axhline(4, 0.1, 0.9, c='r');
```

```python
plt.axvline(2, c='r')
plt.axvline(3, 0.1, 0.8, c='r')
plt.plot(y, 'o-');
```

Une ligne générale est indiqué avec `axline` et deux points.

```python
plt.plot(y, 'o-');
plt.axline((1, 2), (2, 1.5), c='r')
plt.axline((1, 2), (2, 2.5), c='r');
```

## Lignes multiples

```python
plt.plot(y, 'o-')
plt.hlines((4, 4.3, 4.6), 0.5, 4, color='r');
```

```python
plt.plot(y, 'o-')
plt.vlines((1, 1.2, 1.4), 2, 5, color='r');
```

## Bandes

La méthode `axhspan` et `axvspan` permet de placer des bandes horzontales ou verticales. 

```python
plt.axhspan(2, 3, alpha=0.3, color='lime')
plt.plot(y, 'o-');
```

```python
plt.axvspan(2, 3, alpha=0.5, color='pink')
plt.plot(y, 'o-');
```

## Boîte de contour
La boite de contour est optionnel.

```python
plt.subplots(figsize=(8, 2))
plt.plot(y, 'o-');
plt.box(False)
plt.grid(True)
```

## Marques de graduation

```python
plt.plot(y, 'o-')
plt.minorticks_on()
```

```python
plt.plot(y);
plt.box(True)
plt.tick_params(length=10, width=2, direction='inout', 
                color='red', pad=10, labelcolor='g')
```

```python
plt.plot(y);
plt.tick_params(length=0, labelleft=False, labelbottom=False)
```

## Valeurs par défault

Toutes les valeurs par défault sont définis dans **rc** (runtime configuration).

```python
plt.rc('figure', figsize=(7, 2))
plt.rc('lines', linewidth=2, color='r', linestyle=':')
```

```python
plt.plot(y);
```

```python
plt.bar(range(5), y);
```

## Sous-graphes
La fonction `subplot(n, m)` crée $n \times m$ sous-graphes.

```python
plt.subplots();
```

```python
plt.subplots(2, 1);
```

```python
plt.subplots(1, 2);
```

## Spécifier un sous-graphe
La fonction `subplot(n, m, i)` sélectionne le i-ième sous-graphique d'une matrice de $n \times m$ sous-graphes. 

Si le nombre des graphes est inférieur à 10 on peut les désigner directement avec `subplot(nmi)`

```python
x = [0, 1, 2, 3, 4]
y = [1, 3, 2, 5, 2]
```

```python
plt.subplot(121)
plt.plot(y, 'ro-')
plt.subplot(122)
plt.bar(x, y);
```

Ici les graphes sont superposés.

```python
t = np.arange(0.0, 5.0, 0.1)

plt.subplot(211)
plt.plot(t, np.exp(-t) * np.cos(2*np.pi*t), 'bo-')

plt.subplot(212)
plt.plot(t, np.cos(2*np.pi*t), 'r--');
```

## Sous-graphes en 2x2

Nous pouvons aussi sous-diviser en $2 \times 2$ sous-graphes.

```python
plt.subplot(221)
plt.bar(x, y)
plt.subplot(222)
plt.plot(y, 'ro-');
plt.subplot(223)
plt.plot(x, y)
plt.subplot(224)
plt.plot(x, y, 'og');
```

## Arrangement particulier

Il est possible de varier le nombre de sous-graphes dans chaque ligne.

```python
plt.subplot(2,1,1)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,4)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,5)
plt.xticks([]), plt.yticks([])

plt.subplot(2,3,6)
plt.xticks([]), plt.yticks([]);
```

## Enregistrer une figure

La fonction `savefig()` permet de sauvegarder une figure. L'extension du fichier détermine le format.

```python
y = [2, 4, 3, 5, 1]
plt.plot(y, 'o-');
plt.savefig('img/demo.png')
plt.savefig('img/demo.jpg')
plt.savefig('img/demo.pdf')
```

Les trois fichiers se trouvent dans le dossier local. Le PDF est le plus petit, le JPG le plus grand.
