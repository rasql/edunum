## Histogramme

Nous analysons une distribution Gaussienne de 10'000 points.

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
```

```python
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, 
                            facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100$')
plt.text(60, .022, r'$\sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([50, 150, 0, 0.03])
plt.grid(True)
```

## Fr√©quence de lettres

Un dictionnaire est une structure qui permet de compter facilement les lettres dans une phrase.

```python
s = 'Introduction to Python programming'
d = {}
for c in s:
    d[c] = d.get(c, 0) + 1
```

Ceci nous permet de dessiner un histogramme.

```python
plt.bar(d.keys(), d.values());
```

## Histogramme superpos√©

Avec l'option `bottom` nous pouvons superposer un graphique en barre sur un autre. Avec l'option `yerr`nous ajoutons une barre d'erreur.

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std,
        bottom=men_means,
        label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
