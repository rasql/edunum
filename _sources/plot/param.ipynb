# Paramètres

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Taille

La taille par défaut est 6.4 x 4.8 pouces.

```python
plt.subplots(figsize=(10, 2))
```

## Couleur

```python
plt.subplots(facecolor='lightgray', figsize=(8, 2));
```

```python
plt.subplots(figsize=(8, 2), facecolor='pink', 
             edgecolor='red', linewidth=3, frameon=True);
```

```python
y = [2, 4, 3, 5, 1]
plt.subplots(figsize=(8, 2), facecolor='pink', 
             edgecolor='red', linewidth=3, frameon=True)
plt.plot(y, 'o-');
```

```python
plt.subplots(facecolor='pink')
plt.plot(y, 'or-', visible=True);
plt.xlabel('xlabel')
```

```python
plt.plot(y);
plt.arrow(1, 4, 1, 1, shape='full')
plt.axhline(2)
plt.axvline(1)
plt.axhline(4)
```

## Line

```python
plt.axhline(2)
plt.axhline(4)
plt.axhline(4, 1, 2)
plt.plot(y, 'o-');
```

```python
plt.axvline(2)
plt.axvline(3)
plt.plot(y, 'o-');
```

```python
plt.plot(y, 'o-');
plt.axline((1, 2), (2, 1.5))
plt.axline((1, 2), (2, 2.5))
```

## Span

```python
plt.axhspan(2, 3, alpha=0.5)
plt.plot(y, 'o-');
```

```python
plt.axvspan(2, 3, alpha=0.5)
plt.plot(y, 'o-');
```

```python
plt.subplots(figsize=(8, 2))
plt.plot(y, 'o-');
plt.box(False)
plt.grid(True)
```

```python
plt.plot(y, 'o-')
plt.hlines((4, 4.3, 4.6), 1, 4);
plt.hlines(3, 2, 4);
plt.axhline(2.5);
```

```python
plt.plot(y, 'o-')
plt.vlines((1,1.2, 1.4), 2, 5);
```

```python
plt.plot(y, 'o-')
plt.minorticks_off()
```

## Paramètres par défault

```python
plt.rc('figure', figsize=(6, 1))
plt.rc('lines', linewidth=2, color='r', linestyle=':')
```

```python
plt.plot(y);
```

```python
plt.bar(range(5), y);
```

```python
plt.rcdefaults()
plt.plot(y, 'o-');
```

## Paramètres des ticks

```python
plt.plot(y);
plt.box(True)
plt.tick_params(length=10, width=2, direction='inout', color='red', pad=10, labelcolor='g')
```

```python
plt.plot(y);
plt.tick_params(length=0, labelleft=False, labelbottom=False)
```
