# Images

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Charger une image

Pour charger des images nous importons la fonction `imread`.

```python
from matplotlib.image import imread
```

La fonction `imread()` peut importer deux types d'images

- couleur (RGB)
- noir et blancs (0..255)

Voici une image en couleur

```python
img = imread('img/Bachalpsee.jpg')
plt.imshow(img);
```

Voici les dimensions de la matrice:
- 188 pixelx en hauteur
- 250 pixels en largeur
- 3 octets pour dfinir une couleur

```python
img.shape
```

Voici le pixel  la position (25, 25). C'est dans la rgion ciel et de couleur cyan. Les composant bleu et vert sont au maximum

```python
img[25, 25]
```

## Annotation

Nous pouvons superposer du text  une image.

```python
plt.imshow(img);
plt.text(150, 25, 'peak', fontsize=16)
plt.text(100, 110, 'lake', fontsize=16);
```

## Sans les axes

La fonction `axis('off')` permet de ne pas afficher les axes.

```python
plt.imshow(img)
plt.axis('off');
```

## Noir et blancs

Les images noir-et-blancs sont affich avec une **colormap**.

```python
img = imread('img/lake.jpg')
```

```python
plt.imshow(img, cmap='gray');
```

L'image est de taille 220x170 et du type `ndarray`.

```python
img.shape
```

```python
type(img)
```

## Color maps
Les images en noir et blancs peuvent tre coloris avec des colormaps.

```python
maps = ['winter', 'summer', 'autumn', 'gray']

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(img, cmap=maps[i])
    plt.axis('off')
    plt.title(maps[i])
```

## Afficher la barre des couleurs

```python
    plt.imshow(img, cmap='hot')
    plt.axis('off')
    plt.colorbar();
```

## Origin de l'image

Pour inverser l'image nous utilison l'option `origin`.

```python
plt.gray()
plt.imshow(img, origin='lower', cmap='hot');
```

```python
plt.imshow(img, cmap='terrain')
plt.colorbar();
```

## Juxtaposer

```python
a = np.hstack([img, img])
plt.imshow(a);
```

## Superposer

```python
a = np.vstack([img, img])
plt.imshow(a);
```
