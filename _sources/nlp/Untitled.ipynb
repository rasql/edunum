# Processing Raw Text

- access text on the web
- split into words


```python
import nltk, re, pprint
```

```python
from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
```

```python
print(response)
raw = response.read().decode('utf8')
```

```python
type(raw), len(raw)
```

```python
raw[1:78]
```

## Tokenization

```python
from nltk import word_tokenize
```

```python
tokens = word_tokenize(raw)
```

```python
type(tokens)
```

```python
len(tokens)
```

```python
tokens[:10]
```

```python
text = nltk.Text(tokens)
type(text)
```

```python
text[:10]
```

```python
text.collocations()
```

```python
a = raw.find('PART I')
b = raw.find('End of Project Gutenberg')
a, b
```

```python
raw = raw[a:b]
```
