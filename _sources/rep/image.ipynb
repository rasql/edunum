# Images

La grotte de Lascaux est temoin d'une des premières tentatives de l'humanité de figer des images sur un support matériel.
En termes de science d'informatique il s'agit de **l'encodage d'une image** sur un substrat rocheux à l'aide de peinture.

![](https://upload.wikimedia.org/wikipedia/commons/1/1e/Lascaux_painting.jpg)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matrice 2D

Une image est un tableau en 2 dimension composé de petits rectangles. On appelle ces points élémentaires des pixels, de l'anglais **picture element**.

Une image noir et blanc est une image ou chaque pixel peut avoir deux valeurs: 

- 0 = noir
- 1 = blanc

La fonction `np.zeros(m, n)` crée une matrice 2D de zeros de dimension m x n.

```python
img = np.zeros((5, 8), dtype='uint8')
print(img)
```

Nous pouvons afficher une matrice 2D sous forme d'image avec la fonction `np.imshow()`

```python
plt.grid()
plt.imshow(img, cmap='gray');
```

L'attribut `shape` retourne les dimensions de l'image. Ici nous avons 5 lignes et 8 colonnes.

```python
img.shape
```

L'attribut `size` retourne le nombre de pixels. Ici nous avons 40

```python
img.size
```

## Pixels
Les éléments d'une image sont appelé des pixels. Nous pouvons les adresser avec les indices `img[y, x]`. Voici comment changer deux pixels en blanc.

```python
img[1, 2] = 1
img[2, 6] = 1
print(img)
```

Nous pouvons afficher cette image avec 

```python
plt.imshow(img, cmap='gray');
```

## Marice `ones`

La matrice `np.ones()` permet de créer une matrice avec des `1`.
Nous l'utilisons pour créer un **smiley** minimaliste.

```python
img = np.ones((5, 5), dtype='uint8')
img[1, 1] = 0
img[1, 3] = 0
img[3, 1:4] = 0
plt.imshow(img, cmap='gray');
```

## Affichage matricielle

Les chiffres sur un affichage d'une calculatrice sont composé de pixels.

![](img/lcd_display.jpg)

Nous pouvons créer nos propres images d'une lettre avec une matrice 6 x 5 pour créer la lettre **E**.

```python
E = np.array([[1, 0, 0, 0, 1], 
              [1, 0, 1, 1, 1],
              [1, 0, 0, 0, 1],
              [1, 0, 1, 1, 1],
              [1, 0, 0, 0, 1],
              [1, 1, 1, 1, 1]])
plt.imshow(E, cmap='gray')
plt.grid()
```

La dimension de la lettre est bien 6 x 5.

```python
E.shape
```

Voici encore la lettre L

```python
L = np.array([[1, 0, 1, 1, 1], 
              [1, 0, 1, 1, 1],
              [1, 0, 1, 1, 1],
              [1, 0, 1, 1, 1],
              [1, 0, 0, 0, 1], 
              [1, 1, 1, 1, 1]])
plt.imshow(L, cmap='gray')
plt.grid()
```

## Création de texte

Un ordinateur affiche un texte en juxtaposant les pixels d'une lettre.

La fonction `np.hstack()` (=horizontal stack) permet d'empiler des images horizontalement.

```python
ELLE = np.hstack((E, L, L, E))
plt.imshow(ELLE, cmap='gray');
```

La fonction `np.vstack()` permet d'empiler des images verticalement.

```python
img = np.vstack((ELLE, ELLE, ELLE))
plt.imshow(img, cmap='gray');
```

## Lignes

Avec l'opérateur de tranche `:` nous pouvons choisir des lignes entières ou  partielles. 
- `m:n` de m à n
- `:n` du début à n
- `m:` de m jusqu'à la fin

```python
img = np.ones((5, 8), dtype='uint8')
img[0, :] = 0
img[2, :2] = 0
img[3, 2:5] = 0
img[4, 5:] = 0
plt.imshow(img, cmap='gray');
```

## Colonnes

Avec l'opérateur de tranche `:` nous pouvons choisir des lignes ou colonnes entières. 

```python
img = np.ones((5, 8), dtype='uint8')
img[1, :] = 0
img[:, 2] = 0
plt.imshow(img, cmap='gray');
```

## Pixels aléatoires

La fonction `np.random.randint(2)` crée des valeurs aléatoires entre 0 et 1. 

```python
A = np.random.randint(2, size=(5, 10))
plt.imshow(A, cmap='gray');
```

De plus nous pourrions choisir des valeurs aléatoires sur un spectre de gris avec 256 valeurs.

```python
A = np.random.randint(256, size=(5, 6))
plt.imshow(A, cmap='gray')
plt.colorbar();
```
