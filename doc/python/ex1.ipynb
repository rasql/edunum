## Carrés

Quel est le code pour afficher ces 8 lignes de carrés ?  
(2 lignes de code suffisent)

```python
for i in range(8):
    print(i, '** 2 =', i ** 2)
```

## Puissances de 2

Quel est le code pour afficher ces 8 lignes de puissances de 2 ?  
(2 lignes de code suffisent)

```python
for i in range(8):
    print('2 **', i, '=', 2 ** i)
```

## Des carrés sur une lignes

Quel est le code pour afficher ces carrés sur une seule ligne ?

```python
for i in range(1, 21):
    print(i ** 2, end=', ')
```

## Des puissances sur une lignes

Quel est le code pour afficher ces carrés sur une seule ligne ?

```python
for i in range(16):
    print(2 ** i, end=', ')
```

## Table de multiplication

Utilisez deux boucles `for` imbriquées pour créer une table de multiplication.

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(i * j, end='\t')
    print()
```

## Table de vérité `and`

Utilisez deux boucles impriquées qui itèrent sur le tuple `(False, True)`
pour créer une table de verité pour la fonction logique `and`.

```python
print('p \tq \tp and q')
print('--------' * 3)
for p in (False, True):
    for q in (False, True):
        print(p, q, p and q, sep='\t')
```

## Suite de Fibonacci

La suite de Fibonacci est obtenu en additionnant les deux nombres précédents.
Utilisez une boucle pour calculer ses 15 premiers éléments.

```python
a = 0
b = 1
for i in range(15):
    c = a + b
    print(c, end=', ')
    a = b
    b = c
```
