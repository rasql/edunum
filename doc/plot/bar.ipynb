# Diagrammes à barres

Un diagramme à barre est un graphique qui présente des catégories avec des barres rectangulaires. 
La taille du rectangle est proportionnelle à la valeur.
Les barres peuvent être tracés horizontalement ou verticalement.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Les commandes pour barres horizontale/verticale sont similiare:

- `bar(x, height, width=0.8, bottom=None, *, align='center', **kwargs)`
- `barh(y, width, height=0.8, left=None, *, align='center', **kwargs)`



- `bottom` - base de barres
- `aligne` - center/edge

## Vertical
Voici un graph vertical

```python
labels = ['Jim', 'Kim', 'Lee', 'Tim', 'Edy']
a = [2, 4, 3, 5, 1]
plt.bar(labels, a);
plt.title('vertical')
```

## Horizontal
Voici un graph vertical

```python
plt.barh(labels, a)
plt.title('horizontal');
```

```python
errors = [0.5, 0.5, 0.2, 0.7, 0.2]
plt.bar(labels, a, yerr=errors)
plt.title("Avec barre d'erreur");
```

## Epaisseur


```python
plt.bar(labels, a, 0.5, yerr=errors);
```

```python
width = [0.1, 0.2, 0.4, 0.8, 0.2]
plt.bar(labels, a, width, yerr=errors);
```

## Alignement

L'alignement peut être `center` ou `edge`

```python
plt.bar(labels, a, align='edge');
```

## Etiquette

```python
plt.bar(labels, a, yerr=errors, label='A');
plt.legend();
```

## Empilé

```python
b = [1, 2, 3, 2, 1]
plt.bar(labels, a, label='A');
plt.bar(labels, b, bottom = 6, label='B');
plt.legend();
```

```python
plt.bar(labels, a, label='A');
plt.bar(labels, b, bottom = a, label='B');
plt.legend();
```

```python
plt.barh(labels, a, label='A');
plt.barh(labels, b, left = 6, label='B');
plt.legend();
```

```python
plt.barh(labels, a, label='A');
plt.barh(labels, b, left = a, label='B');
plt.legend();
```

## Ajouter un texte

```python
plt.bar(labels, a)
for x, y in enumerate(a):
    plt.text(x, y, y)
```

```python
plt.bar(labels, a)
for x, y in enumerate(a):
    plt.text(x, y/2, y, fontsize=16, color='white', ha='center', va='center')
```

## Groupé

```python
x = np.arange(5)
w = 0.4
```

```python
plt.bar(x-w/2, a, w, label='A')
plt.bar(x+w/2, b, w, label='B')
plt.legend()
plt.xticks(x)
plt.gca().set_xticklabels(labels);
```

```python
plt.barh(x-w/2, a, w, label='A')
plt.barh(x+w/2, b, w, label='B')
plt.legend()
plt.yticks(x)
plt.gca().set_yticklabels(labels);
```
