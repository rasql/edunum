# Diagramme en secteurs

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Pie chart de base
Le graphique basic affiche des tranches.

```python
n = 4
x = range(1, n)
plt.pie(x, labels=x, shadow=True);
```

Cette fois nous avons 10 valeurs avec une distribution uniforme dans l'intervalle $[0, 1]$.

```python
n = 10
Z = np.random.uniform(0, 1, n)
plt.pie(Z, labels=range(n));
```

## Tranches qui ressortent
L'option `explode` permet de faire ressortir les tranches. C'est une liste avec le même nombre d'élément qu'il y a des tranches. La valeur est de 0 à 1 et indique la fraction du rayon pour laquelle la tranche ressort.

```python
plt.pie(Z, labels=range(n), explode=[0.1]*n);
```

Ici seulement la première tranche ressort, d'une valeur de 0.5.

```python
plt.pie(Z, labels=range(n), explode=[0.5] + [0]*(n-1));
```

## Diagramme en escalier

La valeur par défaut est `where='pre'`. La marche commence avec le point.

```python
n = 7
y = range(n)
plt.step(y, 'o-');
```

Avec `where='mid'` la marche est au milieu.

```python
plt.step(y, 'o-', where='mid');
```

```python
plt.step(y, 'o-', where='post');
```

## Données aléatoires

```python
n = 20
x = np.random.uniform(0, 1, n)
y = np.random.uniform(0, 1, n)

plt.plot(x, y, 'o-');
```

```python
plt.step(x, y, 'o-');
```

## Contour plot

```python
def f(x,y): 
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
plt.contour(X, Y, f(X,Y), 8, colors='black');
```
