# Texte

Il est possible d'ajouter du texte et même des formules mathématiques à un graphique.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.text(0.1, 0.1, 'text', fontsize=100);
```

## Ajouter du texte

La fonction 
- `text` ajoute un text dans les axes,
- `figtext` ajoute un text dans la figure.

```python
plt.grid()
plt.text(0.0, 0.2, 'text', fontsize=30)
plt.text(0.4, 0.2, 'text', fontsize=30)
plt.figtext(0, 0, 'figtext', fontsize=20)
plt.figtext(0, 0.5, 'figtext', fontsize=20);
```

## Texte avec bordure

L'objet `Text` a beaucoup d'options.

```python
y = [2, 4, 3, 6, 3]
plt.plot(y, 'o-')
plt.text(1, y[1], 'italic', style='italic', fontsize=20)
plt.text(2, y[2], 'boxed', fontsize=20,
        bbox={'facecolor': 'lime', 'alpha': 0.5, 'pad': 5});
```

## Étiquettes

Matplotlib possède les fonctions suivantes pour placer un titre et des étiquettes pour les axes x et y.

```python
plt.title('Title', fontsize=20)
plt.xlabel('xlabel', fontsize=20)
plt.ylabel('ylabel', fontsize=20);
```

## En position aléatoire

```python
plt.subplot()
for c in 'Python %*+':
    s = np.random.randint(10, 60)
    plt.text(*np.random.rand(2), c, fontsize=s, ha='center')
```

```python
plt.subplot()
for c in 'Python is a nice language'.split():
    plt.text(*np.random.rand(2), c, fontsize=30)
```

## Annotation

## Police de caractère

```python
styles = ['serif', 'sans-serif','cursive', 'fantasy', 'monospace']

for i, s in enumerate(styles):
    plt.text(0.2, 0.05+i/5, s, family=s, fontsize=30)
```

```python
fonts = ['Sans', 'Helvetica', 'Arial', 'Times']
for i, s in enumerate(fonts):
    plt.text(0.2, 0.05+i/4, s, family=s, fontsize=30)
```

## Rotation

L'angle du texte peut être choisi librement.

```python
rotations = [0, 15, 30, 45]
for i, s in enumerate(rotations):
    plt.text(0.2, 0.05 + i/5, str(s)+' deg', rotation=s, fontsize=30)
```

## Poids

```python
weights = ['normal', 'bold', 'heavy', 'light', 'ultralight']
for i, s in enumerate(weights):
    plt.text(0.2, 0.05+i/5, s, weight=s, fontsize=30)
```
