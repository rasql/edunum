### Nombre naturel
Un nombre naturel est reprÃ©sentÃ© avec une sÃ©quence de chiffres.

```python
def n(m):
    s = ''
    for i in range(m):
        s += d()
    return s        
```

Voici trois nombres naturels.

```python
n(3), n(5), n(7)
```

### Nombre entier

Les nombre entiers incluent aussi les nombres nÃ©gatives. Nous utilisons les symboles
- plus `+`
- minus `-`

```python
def z(m):
    sign = choice('-+')
    return  sign + n(m)    
```

Voici trois exemples de nombre entier.

```python
z(2), z(3), z(5)
```

### Nombre rationel

Nombre a virgule flottante

```python
def q(k, m):
    return z(k) + '.' + n(m)
```

```python
q(2, 3), q(1, 7), q(4, 5)
```

### Nombre scientifique

```python
def e(k, m):
    return q(1, k-1) + 'e' + z(m)    
```

```python
e(3, 1), e(2, 2), e(3, 1)
```
