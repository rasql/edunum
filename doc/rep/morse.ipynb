# Code Morse 

Le code Morse permet de transmettre un texte à l'aide d'impulsions courtes et longues.

Inventé en 1832 pour la télégraphie, ce codage de caractères assigne à chaque lettre, chiffre et signe de ponctuation une combinaison unique de signaux.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8e/International_Morse_Code-fr.svg/450px-International_Morse_Code-fr.svg.png)

```python
morse = {'a':'.-', 'b':'-...', 'c':'-.-.', 'd':'-..', 'e':'.', 'f':'..-.', 
        'g':'--.', 'h':'....', 'i':'..', 'j':'.---', 'k':'-.-', 'l':'.-..', 'm':'--',
        'n':'-.', 'o':'---', 'p':'.--.', 'q':'--.-', 'r':'.-.', 's':'...', 't':'-',
        ' ':'  '}
```

```python
for x in morse:
    print(x, '=', morse[x])
```

```python
phrase = 'le code Morse'
```

```python
for c in phrase:
    c = c.lower()
    print(morse[c], end='  ')
```

## Créer un son

```python
import simpleaudio as sa
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 0.1, 800)
audio = 10000 * np.sin(t * 3000)
plt.plot(audio)
audio = audio.astype(np.int16)
sa.play_buffer(audio, 1, 2, 16000)
```

```python
audio = 30000 * np.sin(t * 2000)*t
audio = audio.astype(np.int16)
sa.play_buffer(audio, 1, 2, 16000);
```

```python
from time import sleep
```

```python
for i in range(3):
    sa.play_buffer(audio, 1, 2, 16000);
    sleep(0.1)
```

```python
t = np.linspace(0, 0.3, 2400)
da = 10000 * np.sin(t * 3000)
plt.plot(da)
da = da.astype(np.int16)
sa.play_buffer(da, 1, 2, 16000)
```

```python
t = np.linspace(0, 0.1, 800)
di = 10000 * np.sin(t * 3000)
plt.plot(di)
di = di.astype(np.int16)
sa.play_buffer(di, 1, 2, 16000)
sleep(0.3)
sa.play_buffer(da, 1, 2, 16000)
```

```python
for c in phrase:
    c = c.lower()
    print(morse[c], end='  ')
    for s in morse[c]:
        if s == '.':
            sa.play_buffer(di, 1, 2, 16000)
        elif s == '-':
            sa.play_buffer(da, 1, 2, 16000)
        sleep(0.2)
    sleep(0.2)
```
