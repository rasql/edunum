# Couleur
Toutes les couleurs peuvent être créés avec les couleurs de base rouge, vert et bleu.

![](https://upload.wikimedia.org/wikipedia/commons/1/14/AdditiveColorMixing.png)

Le mélange des 3 couleurs donne la couleur blanche.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Couleurs de base

Le système RGB (Red, Green Blue) définit une couleur par ces 3 composantes.
Une couleur est donc un tuple, et nous pouvons définir les 3 coulers de base comme ceci:

```python
R = np.array([255, 0, 0])
G = np.array([0, 255, 0])
B = np.array([0, 0, 255])
```

Ce nous permet de créer une image avec 3 pixels: R, G, B

```python
img = np.array([[R, G, B]])
plt.imshow(img);
```

## Mélanger des couleurs

Les couleurs sont additives.

- R+B donne **magenta** (M)
- R+G donne **jaune** (Y)
- G+B donne **cyan** (C)

```python
M = R + B
Y = R + G
C = B + G
```

Vérifions les couleurs

```python
print(Y, 'yellow')
print(C, 'cyan')
print(M, 'magenta')
```

```python
img = np.array([[M, C, Y]])
plt.imshow(img);
```

## Noir et blanc

Toutes les couleurs mélangées donnent blanc (W=white). L'absence de toute couleur donne noir (K=black).

```python
W = R + G + B
K = [0, 0, 0]
```

```python
img = np.array([[M, W, C, K, Y]])
plt.imshow(img);
```

Nous pouvons maintenant créer des simples icones, comme par exemple le drapeau Suisse

```python
ch = np.array([[R, R, R, R, R],
               [R, R, W, R, R],
               [R, W, W, W, R],
               [R, R, W, R, R],
               [R, R, R, R, R]])
plt.imshow(ch);
```

## Saturation.

Une couleur avec sa composante au maximum (255) est dit saturée.
Elle peut être rendue plus foncé en diminuant cette composante. Voici 4 valeurs de saturation (100%, 50%, 33%, 25%)

```python
img = np.array([[R, R//2, R//3, R//4]])
plt.imshow(img);
```

Voici des couleurs qui s'approchent du blanc.

```python
img = np.array([[R, R+(G+B)//3, R+(G+B)//2, R+3*(G+B)//4]])
plt.imshow(img);
```

## Charger une image

La fonction `plt.imgread()` permet de charger une image.

```python
img = plt.imread('img/fleurs.jpg')
plt.imshow(img);
```

Trouvons sa dimension

```python
img.shape
```

C'est une image avec 256x256 pixels et 3 composantes couleurs RGB.

## Multiplier et diviser

Nous pouvons multiplier et diviser les matrices **numpy**. Divisées par 2, l'image devient plus sombre.

```python
plt.imshow(img//2);
```

Multipliées par 2, l'image devient plus clair et montre des traces de saturation.

```python
plt.imshow(img*2);
```
